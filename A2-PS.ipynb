<h1>Regiões e Estados Brasileiros</h1>

```python
def IBGE(x):
    #array de cada instância de estado

    ibge = x.cd_ibge.array
    estados = []
    for i in range(len(ibge)):
      estados.append(int(str(ibge[i])[:2]))

    #contador de ocorrências de cada estado

    estados_unicos = []
    count_estados = []

    for estado in estados:
        if (estado not in estados_unicos): #verifica se o estado atual não está no array de estados
            estados_unicos.append(estado)
            count_estados.append(estados.count(estado))

    #array de cada entrada de regiões
    regioes = []
    for i in range(len(estados)):
      regioes.append(int(str(estados[i])[:1]))

    #contador de ocorrências de cada região
    count_regioes = [regioes.count(1), regioes.count(2), regioes.count(3), regioes.count(4), regioes.count(5)]
    
    return estados_unicos, count_estados, regioes, count_regioes
```

```python
#gera os dados de regiões e estados para o cadúnico 2018
estados_unicos, count_estados, regioes, count_regioes = IBGE(bf)
```
