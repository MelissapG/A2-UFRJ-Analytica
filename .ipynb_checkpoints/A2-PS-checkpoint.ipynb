<h1>Análise de Dados Desidentificados do Cadastro Único (2012 e 2018)</h1>

```python
#cadastro único 2018
bf2018_fp = 'base_amostra_pessoa_201812.csv'
bf=pd.read_csv(bf2018_fp, encoding='utf-8', sep = ';', dtype={'id_pessoa': 'string'})
```

```python
#cadastro único 2012
bf2012_fp = 'base_amostra_pessoa_201212.csv'
bf2=pd.read_csv(bf2012_fp, encoding='utf-8', sep = ';', dtype={'id_pessoa': 'string'})
```

<h1>Série de Pessoas</h1>

```python
#Serie de pessoas entre 18 e 65 anos 
z = bf.loc[bf.idade.between(18,65, inclusive=True)]
count = z.idade.value_counts()
```

```python
#Serie de pessoas entre 15 e 65 anos - PEA - população economicamente ativa
pea = bf.loc[bf.idade.between(15,65, inclusive=True)]
count_pea = pea.idade.value_counts()
```

<h1>Pessoas</h1>

```python
fig, ax = plt.subplots(1,2, figsize=(30,10))
ax[0].title.set_text('Pessoas inscritas no Cadastro Único - 2012')
ax[1].title.set_text('Pessoas inscritas no Cadastro Único - 2018')
sns.histplot(bf2.idade, color='r', ax=ax[0])
sns.histplot(bf.idade, color='b', ax=ax[1])
plt.show()
```

```python
print("Total inscritos no CADÚnico 2012: {}".format(bf2.idade.value_counts().sum()))
print("Total inscritos no CADÚnico 2018: {}\n".format(bf.idade.value_counts().sum()))
```

```python
idadesbf = pd.cut(bf.idade, [0, 20, 60, 130], labels=["jovem", "adulto", "idoso"]) #reunindo as idades em 3 grupos
idadesbf2 = pd.cut(bf2.idade, [0, 20, 60, 130], labels=["jovem", "adulto", "idoso"]) #reunindo as idades em 3 grupos

fig, axs = plt.subplots(1, 2,figsize=(16,10))
fig.suptitle("Grupo de Idade 2012 - 2018", fontsize=18)
plt.subplots_adjust(top=1.1)
ax[0].title.set_text('Grupo de Idade - 2012')
ax[1].title.set_text('Grupo de Idade - 2018')
axs[0].pie(idadesbf2.value_counts(), labels=['Adulto','Jovem','Idoso'], autopct='%1.1f%%', shadow=True)
axs[1].pie(idadesbf.value_counts(), labels=['Adulto','Jovem','Idoso'], autopct='%1.1f%%', shadow=True)
plt.show()
```

<h1>Regiões e Estados Brasileiros que mais utilizam o Bolsa Família</h1>

```python
#array de cada instância de estado

ibge = bf.cd_ibge.array
estados = []
for i in range(len(ibge)):
  estados.append(int(str(ibge[i])[:2]))
```

```python
#contador de ocorrências de cada estado

estados_unicos = []
count_estados = []

for estado in estados:
    if (estado not in estados_unicos): #verifica se o estado atual não está no array de estados
        estados_unicos.append(estado)
        count_estados.append(estados.count(estado))
```

```python
#array de cada entrada de regiões
regioes = []
for i in range(len(estados)):
  regioes.append(int(str(estados[i])[:1]))

#contador de ocorrências de cada região
count_regioes = [regioes.count(1), regioes.count(2), regioes.count(3), regioes.count(4), regioes.count(5)]
```

```python
plt.figure(figsize=(10,6))
plt.title("Regiões e o Bolsa Família")
plt.pie(count_regioes, labels=["Norte", "Nordeste", "Sudeste", "Sul", "Centro-Oeste"], autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(22,10))
plt.title("Estados e o Bolsa Família")
sns.barplot(y=count_estados, x=estados_unicos)
plt.show()
```
