<h1>Regiões e Estados Brasileiros</h1>

```python
def IBGE(x):
    #array de cada instância de estado

    ibge = x.cd_ibge.array
    estados = []
    for i in range(len(ibge)):
      estados.append(int(str(ibge[i])[:2]))

    #contador de ocorrências de cada estado

    estados_unicos = []
    count_estados = []

    for estado in estados:
        if (estado not in estados_unicos): #verifica se o estado atual não está no array de estados
            estados_unicos.append(estado)
            count_estados.append(estados.count(estado))

    #array de cada entrada de regiões
    regioes = []
    for i in range(len(estados)):
      regioes.append(int(str(estados[i])[:1]))

    #contador de ocorrências de cada região
    count_regioes = [regioes.count(1), regioes.count(2), regioes.count(3), regioes.count(4), regioes.count(5)]
    
    return estados_unicos, count_estados, regioes, count_regioes
```

```python
#gera os dados de regiões e estados para o cadúnico 2018
estados_unicos, count_estados, regioes, count_regioes = IBGE(cadunico18)
```

```python
#Retorna a Sigla de cada estado em relação ao código
def switch(number):
    switcher = {
        12: "AC",
        27: "AL",
        16: "AP",
        13: "AM",
        29: "BA",
        23: "CE",
        53: "DF",
        32: "ES",
        52: "GO",
        21: "MA",
        51: "MT",
        50: "MS",
        31: "MG",
        15: "PA",
        25: "PB",
        41: "PR",
        26: "PE",
        22: "PI",
        24: "RN",
        43: "RS",
        33: "RJ",
        11: "RR",
        14: "RO",
        42: "SC",
        35: "SP",
        28: "SE",
        17: "TO"
    }
    return switcher.get(number, "")

#Modifica o array de códigos para siglas de estados 
def changeStateNumberForInitials(estados):
    state = []
    for st in estados:
        state.append(switch(st))
    return state
estados_unicos = changeStateNumberForInitials(estados_unicos)        
```

<h1>Renda</h1>

```python
#valor de remuneração media e renda média anual 2012

rem_media2 = pea2.val_remuner_emprego_memb.mean()
renda_media_anual2 = p2.val_renda_bruta_12_meses_memb.mean()
print("Remuneração Média: R${:.2f} e Renda Média Anual: R${:.2f} de 2012".format(rem_media2,renda_media_anual2))

#valor de remuneração media e renda média anual 2018

rem_media = pea.val_remuner_emprego_memb.mean()
renda_media_anual = p.val_renda_bruta_12_meses_memb.mean()
print("Remuneração Média: R${:.2f} e Renda Média Anual: R${:.2f} de 2018".format(rem_media,renda_media_anual))

var_perc_rem = ((rem_media-rem_media2)/rem_media2)*100
var_perc_renda = ((renda_media_anual-renda_media_anual2)/renda_media_anual2)*100

print("\nAumento percentual da Remuneração (PEA): {:.2f}%\nAumento Percentual da Renda Anual: {:.2f}%".format(var_perc_rem, var_perc_renda))
```
